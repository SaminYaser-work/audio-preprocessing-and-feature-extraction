```python
import librosa
import numpy as np
import pandas as pd
import librosa as lr
import librosa.display as lrd
import matplotlib.pyplot as plt
import seaborn as sb
from vscode_audio import Audio
# from IPython.display import Audio
import os
import warnings
warnings.filterwarnings("ignore")

dataset = '..\\tess'
selected_set1 = 'YAF'
selected_set2 = 'OAF'

files = []
emo = []

for dirpath, dirnames, filenames in os.walk(dataset):
    for filename in filenames:
        if selected_set1 in dirpath or selected_set2 in dirpath:
            x = os.path.join(dirpath, filename)
            files.append(x)
            y = x.split('_')[-1].split('.')[0].lower()
            emo.append(y)
```

```python
d = pd.DataFrame()
d['sound'] = files
d['emotion'] = emo

d.head()
```

```python
sb.countplot(x='emotion', data=d)
```

```python
def waveplot(data, sr, title):
    plt.figure(figsize=(12, 4))
    plt.title(title)
    lrd.waveshow(data, sr=sr)
    plt.show()

def spectograph(data, sr, title):
    x = lr.stft(data)
    xdb = lr.amplitude_to_db(abs(x))
    plt.figure(figsize=(12, 4))
    plt.title(title)
    lrd.specshow(xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()

def getPath(type):
    return np.array(d['sound'][d['emotion'] == type])[1]

def getInfo(path, type):
    data, sr = lr.load(path)
    waveplot(data, sr, type)
    spectograph(data, sr, type)
```

# Feature Extraction

```python
def mfcc(filename):
    y, sr = lr.load(filename, duration=3, offset=0.5)
    mfcc = np.mean(lr.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
    return mfcc
```

```python
mfcc(d['sound'][0])
```

```python
X_mfcc = d['sound'].apply(lambda x: mfcc(x))
```

```python
X_mfcc
```

```python
X = [x for x in X_mfcc]
X = np.array(X)
X.shape
```

# Input split

```python
X = np.expand_dims(X, -1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(d[['emotion']])
```

```python
y = y.toarray()
```

# LTSM

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(123, return_sequences=False, input_shape=(40, 1)),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

# Training

```python
history = model.fit(X, y, validation_split=0.2, epochs=100, batch_size=512, shuffle=True)
```

```python
epochs = list(range(100))
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.plot(epochs, acc, label='train accuracy')
plt.plot(epochs, val_acc, label='val accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs, loss, label='train loss')
plt.plot(epochs, val_loss, label='val loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```
